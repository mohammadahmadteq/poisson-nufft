```python
import numpy as np
import time
import matplotlib.pyplot as plt
import finufft

# Official qc-grid library imports
from grid.onedgrid import GaussLegendre
from grid.rtransform import BeckeRTransform, LinearInfiniteRTransform
from grid.angular import AngularGrid
from grid.atomgrid import AtomGrid
from grid.molgrid import MolGrid
from grid.becke import BeckeWeights

# Mock molecule and basis set for demonstration purposes
# In a real scenario, you'd extract these from PySCF, Psi4, or your own arrays
class MockMolecule:
    def __init__(self):
        self.atomic_numbers = np.array([1, 1])  # H2 molecule
        self.coordinates = np.array([[0.0, 0.0, -0.7], [0.0, 0.0, 0.7]])

mol = MockMolecule()

def mock_basis(points, alpha=1.0):
    # A simple s-type Gaussian for testing: exp(-alpha * r^2)
    r_sq = np.sum(points**2, axis=1)
    return np.exp(-alpha * r_sq), None

basis_set = [mock_basis] * 4
```

```python
def generate_k_grid_qcgrid(n_rad=50, k_max=15.0, angular_degree=11):
    """
    Constructs a 3D k-space grid using qc-grid's Gauss-Legendre and Lebedev tools.
    """
    # 1. Radial Grid on [0, k_max]
    gl = GaussLegendre(npoints=n_rad)
    # Map [-1, 1] to [0, k_max]
    r_pts = (gl.points + 1.0) * (k_max / 2.0)
    r_dr_wts = gl.weights * (k_max / 2.0)
    
    # 2. Angular Grid: Lebedev-Laikov on unit sphere
    ang_grid = AngularGrid(degree=angular_degree)
    ang_pts = ang_grid.points
    ang_wts = ang_grid.weights
    
    n_ang = len(ang_pts)
    n_total = n_rad * n_ang
    
    kx, ky, kz = np.zeros(n_total), np.zeros(n_total), np.zeros(n_total)
    k_mag, k_weights = np.zeros(n_total), np.zeros(n_total)
    
    idx = 0
    for i in range(n_rad):
        r = r_pts[i]
        r_wt = r_dr_wts[i]
        for j in range(n_ang):
            kx[idx] = r * ang_pts[j, 0]
            ky[idx] = r * ang_pts[j, 1]
            kz[idx] = r * ang_pts[j, 2]
            k_mag[idx] = r
            # Measure: d^3k = r^2 * dr * d(Omega)
            k_weights[idx] = (r**2) * r_wt * ang_wts[j]
            idx += 1
            
    return kx, ky, kz, k_mag, k_weights
```

```python
def compute_eri_nufft(mgrid, basis, kx, ky, kz, k_mag, k_weights):
    x, y, z = mgrid.points[:,0], mgrid.points[:,1], mgrid.points[:,2]
    w_r = mgrid.weights
    
    # 1. Overlap density g(r)
    g_r = basis[0](mgrid.points)[0] * basis[1](mgrid.points)[0]
    c_forward = (g_r * w_r).astype(complex)
    
    # 2. Forward NUFFT: g(r) -> G(k)
    tol = 1e-8
    G_k = finufft.nufft3d3(x, y, z, c_forward, kx, ky, kz, isign=-1, eps=tol)
    
    # 3. Poisson Equation in k-space: V(k) = (4*pi / k^2) * G(k)
    k_sq = k_mag**2
    k_sq[k_sq < 1e-12] = 1e-12 # Prevent origin singularity
    V_k = (4 * np.pi / k_sq) * G_k
    c_inverse = V_k * k_weights / (2 * np.pi)**3
    
    # 4. Inverse NUFFT: V(k) -> V(r)
    V_r = finufft.nufft3d3(kx, ky, kz, c_inverse, x, y, z, isign=1, eps=tol).real
    
    # 5. Final Integration with f(r)
    f_r = basis[2](mgrid.points)[0] * basis[3](mgrid.points)[0]
    return np.sum(f_r * V_r * w_r)

def compute_eri_real_space(mgrid, basis):
    r1 = mgrid.points
    w1 = mgrid.weights
    
    g_r1 = basis[0](r1)[0] * basis[1](r1)[0]
    f_r2 = basis[2](r1)[0] * basis[3](r1)[0]
    
    eri = 0.0
    for i, r_1 in enumerate(r1):
        dist = np.linalg.norm(r1 - r_1, axis=1)
        dist[dist < 1e-12] = 1e-12 
        
        potential = np.sum((f_r2 * w1) / dist)
        eri += g_r1[i] * potential * w1[i]
        
    return eri
```

```python
grid_scalings = [1,2,3,4,5, 6] # Radial points dictating grid resolution
total_r_points = []
nufft_times, real_times = [], []
nufft_results, real_results = [], []

print("Starting convergence tests... (This may take a moment)")

for n_rad in grid_scalings:
    # Build Molecular Grid in Real Space using qc-grid
    oned = GaussLegendre(npoints=n_rad)
    rg = BeckeRTransform(0.0, R=1.5).transform_1d_grid(oned)
    
    temp_mgrid = MolGrid.from_preset(
        atnums=mol.atomic_numbers, 
        atcoords=mol.coordinates,
        rgrid=rg, 
        preset="coarse", # Keeps angular degree manageable for testing
        aim_weights=BeckeWeights()
    )
    total_r_points.append(temp_mgrid.size)
    
    # Build K-Space Grid using qc-grid
    kx, ky, kz, k_mag, k_weights = generate_k_grid_qcgrid(n_rad=n_rad, k_max=15.0, angular_degree=11)
    
    # Run NUFFT Method
    t0 = time.time()
    res_nufft = compute_eri_nufft(temp_mgrid, basis_set, kx, ky, kz, k_mag, k_weights)
    nufft_times.append(time.time() - t0)
    nufft_results.append(res_nufft)
    
    # Run Regular Iteration (Real Space)
    t0 = time.time()
    res_real = compute_eri_real_space(temp_mgrid, basis_set)
    real_times.append(time.time() - t0)
    real_results.append(res_real)
    
    print(f"Grid points: {temp_mgrid.size} | NUFFT ERI: {res_nufft:.6f} | Real Space ERI: {res_real:.6f}")

# --- Generate the Charts and Graphs ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Convergence Chart
ax1.plot(total_r_points, nufft_results, marker='o', label='NUFFT (finufft)', color='blue')
ax1.plot(total_r_points, real_results, marker='s', label='Standard Iteration', color='red', linestyle='--')
ax1.set_title('Integration Accuracy Convergence')
ax1.set_xlabel('Number of Grid Points (Real Space)')
ax1.set_ylabel('Integral Value')
ax1.legend()
ax1.grid(True)

# Time Comparison Chart
ax2.plot(total_r_points, nufft_times, marker='o', label='NUFFT (finufft)', color='blue')
ax2.plot(total_r_points, real_times, marker='s', label='Standard Iteration', color='red', linestyle='--')
ax2.set_title('Computational Cost')
ax2.set_xlabel('Number of Grid Points (Real Space)')
ax2.set_ylabel('Time (seconds)')
ax2.set_yscale('log') # Log scale is crucial to see O(N log N) vs O(N^2)
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
